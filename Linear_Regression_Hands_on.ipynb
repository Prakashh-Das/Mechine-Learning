```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
np.random.seed(42)

num_samples = 500

years_of_experience = np.random.randint(2, 21, size=num_samples)

slope = (200_000 - 60_000)/18
intercept = 60_000

salary = slope * years_of_experience + intercept + np.random.normal(0, 10_000, size=num_samples)

data = pd.DataFrame({'YearsOfExperience': years_of_experience, 'Salary': salary})

df = pd.DataFrame(data)
```

```python
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='YearsOfExperience', y='Salary', color='blue')
sns.regplot(data=df, x='YearsOfExperience', y='Salary', scatter=False, color='red')

plt.title('Relationship between Years of Experience and Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
x = df[['YearsOfExperience']]
y = df['Salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
coefficient = model.coef_
intercept = model.intercept_
```

```python
x = np.linspace(0, 20, 100)
y = coefficient * x + intercept

plt.scatter(x, y, label=f'y = {coefficient[0]:.2f}x + {intercept:.2f}')
plt.title('Relationship between Years of Experience and Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```
